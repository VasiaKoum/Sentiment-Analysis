## <a href='#0'> 1. Cleaning the data: Preprocessing and Analysis </a>

```python
from preprocess import *
from string import punctuation
from nltk.corpus import stopwords
from collections import Counter
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

np.set_printoptions(threshold=np.inf) # print entire arrays without truncation

%matplotlib inline

text = open('./twitter_data/train2017.tsv', 'r', encoding="utf8").read()
tweets_list = text.splitlines() # split the tweets in lines

# emoticons dictionary
EMOTICONS  = { ":)":"smiley", ":-)":"smiley", ":]":"smiley", ":-]":"smiley", ":3":"smiley", ":-3":"smiley", 
              ":>":"smiley", ":->":"smiley", ":o)":"smiley", "=))":"smiley", "=)":"smiley",
              ":D":"laugh", ":-D":"laugh",
              ":')":"happy",
              ":P":"cheeky",
              "<3":"love",
              "</3":"sad", ":(":"sad", ":-(":"sad", ":c":"sad", ":<":"sad", ":'(":"sad", 
              ":-||":"angry", ">:[":"angry", ":{":"angry", ":@":"angry", ">:(":"angry" 
             }

all_words = []
positive_words = []
negative_words = []
neutral_words = []
all_tweets = []
positive_cnt = 0
negative_cnt = 0
neutral_cnt = 0

for tweet in tweets_list:  # for each tweet
    cur_tweet = []
    words = tweet.split() # split the tweet into a list of words
    reformed = [EMOTICONS[word] if word in EMOTICONS else word for word in words]   # replace every emoticon in the tweet(if any) with the word from the dictionary
    new_tweet = " ".join(reformed)  # reconstruct the tweet
    words = new_tweet.split()       # split the new tweet into a list of words
    if words[2] == 'positive':      # if label = 'positive' increment positive tweets
        positive_cnt = positive_cnt + 1
    elif words[2] == 'negative':    # if label = 'negative' increment negative tweets
        negative_cnt = negative_cnt + 1
    else:                           # if label = 'neutral' increment neutral tweets
        neutral_cnt = neutral_cnt + 1
        
    for word in words[3:]:  # actual tweet starts at index 3
        if (word[0] == '@') or (word[0] == '#') or (word[:4] == 'http'): # ignore @somename, #somehashtag and http://somelink
            continue
        clean_word = word.strip(punctuation).lower() # remove the special characters and then convert it to lowercase
        if clean_word == '': # if the string, after cleaning, is the empty string, ignore it
            continue
        if clean_word in stopwords.words('english'): # ignore the stopwords
            continue
            
        all_words.append(clean_word) # add the cleaned word to the list containing all the words
        # also add the word at the appropriate list
        if words[2] == 'positive':
            positive_words.append(clean_word)
        elif words[2] == 'negative':
            negative_words.append(clean_word)
        else:
            neutral_words.append(clean_word)
            
        cur_tweet.append(clean_word) # add all the cleaned words of the current tweet to a list 
        
    cleaned_tweet = " ".join(cur_tweet) # reconstruct the cleaned tweet
    all_tweets.append(cleaned_tweet)    # append the cleaned tweet to the list of all_tweets
```

## <a href='#0'> 3. Vectorization </a>

<em>BAG-OF-WORDS</em>

```python
from sklearn.feature_extraction.text import CountVectorizer
import pickle

cleaned_tweets_test = clean_tweets('./twitter_data/test2017.tsv', EMOTICONS)  # returns a list of the cleaned tweets - test dataset

bow_vectorizer = CountVectorizer(max_df=0.9, min_df=2, max_features=1000, stop_words='english')
bow_xtrain = bow_vectorizer.fit_transform(all_tweets)          # build the vectors for the training dataset
bow_xtest  = bow_vectorizer.fit_transform(cleaned_tweets_test) # build the vectors for the test dataset

vectors_array = bow_xtrain.toarray()

with open("Bag-of-words.pkl", "wb") as bow:   #Pickling
    pickle.dump(vectors_array, bow)
bow.close()
```

<em>TF-IDF</em>

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words='english')
tfidf_xtrain = tfidf_vectorizer.fit_transform(all_tweets)         # build the vectors for the training dataset
tfidf_xtest = tfidf_vectorizer.fit_transform(cleaned_tweets_test) # build the vectors for the training dataset

vectors_array = tfidf_xtrain.toarray()

with open("Tf-idf.pkl", "wb") as Tf:   #Pickling
    pickle.dump(vectors_array, Tf)
Tf.close()
```

<em>Word embeddings</em>
<li> Model training </li>

```python
import pandas as pd
import gensim

# Word2Vec 
tokenized_tweets_train = [ cleaned_tweet.split() for cleaned_tweet in all_tweets ]
tokenized_tweets_test  = [ cleaned_tweet.split() for cleaned_tweet in cleaned_tweets_test ]
tokenized_tweets = tokenized_tweets_train + tokenized_tweets_test # merge the 2 lists

# train the w2vec model with the words from both the training and test sets
features = 300
model_w2v = gensim.models.Word2Vec(size = features, min_count = 2) 
model_w2v.build_vocab(tokenized_tweets)
model_w2v.train(tokenized_tweets, total_examples=len(all_tweets) + len(cleaned_tweets_test), epochs=20)
model_w2v.save('word2vec.model')
```

<li> Add features</li>

<b> Including: affin and valence_tweet lexicons</b> <br>
Create dictionaries from the texts lexicons.

```python
lexicon_1 = {}
file = open('./lexica/affin/affin.txt', 'rt')
for lines in (file.readlines()):
    add_key = lines.split()[0]
    if lines.split()[0] != lines.split()[-2]:
        add_key = add_key + " " + lines.split()[-2]
    lexicon_1[add_key] = int(lines.split()[-1])

file = open('./lexica/emotweet/valence_tweet.txt', 'rt', encoding="utf8")
lexicon_2 = {lines.split()[0]:float(lines.split()[-1]) for lines in (file.readlines())}
```

VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social media. VADER uses a combination of A sentiment lexicon is a list of lexical features (e.g., words) which are generally labelled according to their semantic orientation as either positive or negative. <br>
Requires the follow command: <b> pip install vaderSentiment </b> <br>
In  <em> preprocess.py </em>  is being used the sentiment_analyzer_scores function where input is the uncleaned,original tweets and the output is the compound score (from positive, negative and neutral score) of the tweets 

```python
# the original tweets (the training and the testing set)
original_train_tweets = [(tweet.split("\t"))[3:] for tweet in tweets_list]
tweets_list = open('./twitter_data/test2017.tsv', 'r').read().splitlines()
original_test_tweets = [(tweet.split("\t"))[3:] for tweet in tweets_list]

model_w2v = gensim.models.Word2Vec.load("word2vec.model") # load the model

# now create a vector for each tweet, calculating the mean of all the word vectors in each tweet
w2v_xtrain = np.concatenate( [tweet2vector(model_w2v, tokens, features, lexicon_1, lexicon_2) for tokens in tokenized_tweets_train] )
w2v_xtrain = vader_sentiment_analysis(w2v_xtrain, original_train_tweets) # returns the compound score

w2v_xtest  = np.concatenate( [tweet2vector(model_w2v, tokens, features, lexicon_1, lexicon_2) for tokens in tokenized_tweets_test] )
w2v_xtest = vader_sentiment_analysis(w2v_xtest, original_test_tweets) # returns the compound score
```

## <a href='#0'> 4. TSNE model </a>

```python
from sklearn.manifold import TSNE

def tsne_plot(model): # Creates and TSNE model and plots it
    labels = []
    tokens = []
    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=1000, random_state=23)
    new_values = tsne_model.fit_transform(tokens)
    
    x = []
    y = []
    
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
    plt.figure(figsize=(16, 16))
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i], xy=(x[i], y[i]), xytext=(5, 2), textcoords='offset points', ha='right', va='bottom')
    plt.show()
# prepare a "smaller" word2vec model
model_w2v_demo = gensim.models.Word2Vec(size = features, min_count = 2)
model_w2v_demo.build_vocab(tokenized_tweets_train[:600])
model_w2v_demo.train(tokenized_tweets_train[:600], total_examples=len(all_tweets[:500]), epochs=20)
tsne_plot(model_w2v_demo)
#model_w2v = gensim.models.Word2Vec.load("word2vec.model") # load the model
#tsne_plot(model_w2v)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report

train_labels = get_labels("./twitter_data/train2017.tsv")          # returns a list of the correct labels - training dataset
correct_labels = get_labels("./twitter_data/SemEval2017_task4_subtaskA_test_english_gold.txt") # returns a list of the correct labels - test dataset

# Bag of Words
knn = KNeighborsClassifier(n_neighbors = 5) 
knn.fit(bow_xtrain, train_labels)              # "feed" the model the training dataset along with the correct labels
y_pred = knn.predict(bow_xtest)                # predicts the labels for the test dataset
f1_score_knn_bow = f1_score(correct_labels, y_pred, average = 'macro')
classification_report_knn_bow = classification_report(correct_labels, y_pred, labels=['positive', 'negative', 'neutral'])

# TF-IDF
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(tfidf_xtrain, train_labels)
y_pred = knn.predict(tfidf_xtest) # predict
f1_score_knn_tf = f1_score(correct_labels, y_pred, average = 'macro')
classification_report_knn_tf = classification_report(correct_labels, y_pred, labels=['positive', 'negative', 'neutral'])

# Word2Vec
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(w2v_xtrain, train_labels)
y_pred = knn.predict(w2v_xtest) # predict
f1_score_knn_w2v = f1_score(correct_labels, y_pred, average = 'macro')
classification_report_knn_w2v = classification_report(correct_labels, y_pred, labels=['positive', 'negative', 'neutral'])
```
